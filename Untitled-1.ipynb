```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
from ta import add_all_ta_features
from ta.utils import dropna
from keras.layers import Input
```

```python

# Step 1: Load and Preprocess Data
start_date = '2006-01-01'
end_date = '2024-05-13'
ticker = 'EURUSD=X'
data = yf.download(ticker, start=start_date, end=end_date)
```

```python
data = data[['Open', 'High', 'Low', 'Close', 'Volume']]
data
```

```python
# Step 2: Feature Engineering
data = add_all_ta_features(data, open="Open", high="High", low="Low", close="Close", volume="Volume")
data
```

```python
# Drop rows with missing values after adding technical indicators
# data.dropna(inplace=True)
data.fillna(0, inplace=True)  # Replace NaN with zero
data
```

```python
# Step 3: Prepare Data for LSTM
# Identify and handle infinity or very large values
data.replace([np.inf, -np.inf], np.nan, inplace=True)
data.fillna(method='ffill', inplace=True)
data.fillna(method='bfill', inplace=True)
```

```python
data
```

```python
# Step 3: Prepare Data for LSTM
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)
```

```python
# Create sequences
def create_sequences(data, seq_length):
    sequences = []
    for i in range(len(data) - seq_length):
        sequences.append(data[i:i+seq_length])
    return np.array(sequences)

seq_length = 60
X = create_sequences(scaled_data, seq_length)
y = scaled_data[seq_length:, 3]  # Predicting the close price
```

```python
# Split data into training and test sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
from keras.layers import Input

# Step 4: Build LSTM Model
model = Sequential()
model.add(LSTM(units=64, return_sequences=True, input_shape=(seq_length, X.shape[2])))
model.add(Dropout(0.2))
model.add(LSTM(units=64))
model.add(Dropout(0.2))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Step 5: Train the Model
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)
```

```python
# Step 6: Make Predictions
predictions = model.predict(X_test)
predictions = scaler.inverse_transform(np.concatenate((np.zeros((predictions.shape[0], scaled_data.shape[1] - 1)), predictions), axis=1))[:, -1]

```

```python
# Plot predictions
plt.figure(figsize=(14, 5))
plt.plot(data.index[train_size + seq_length:], predictions, color='red', label='Predicted Price')
plt.plot(data.index[train_size + seq_length:], data['Close'].values[train_size + seq_length:], color='blue', label='Actual Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

```

```python
# Step 7: Save and Deploy the Model
model.save('forex_lstm_model.h5')
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(data['Close'].values[train_size + seq_length:], predictions)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(data['Close'].values[train_size + seq_length:], predictions)

print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Absolute Error (MAE):", mae)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
from ta import add_all_ta_features
from ta.utils import dropna

# Step 1: Load and Preprocess Data
start_date = '2010-01-01'
end_date = '2024-05-13'
ticker = 'EURUSD=X'
data = yf.download(ticker, start=start_date, end=end_date)

# Ensure data is in the correct format
data = data[['Open', 'High', 'Low', 'Close', 'Volume']]

# Drop rows with missing values
# data.dropna(inplace=True)
data.fillna(0, inplace=True)  # Replace NaN with zero

# Step 2: Feature Engineering
data = add_all_ta_features(data, open="Open", high="High", low="Low", close="Close", volume="Volume")

# Drop rows with missing values after adding technical indicators
# data.dropna(inplace=True)
data.fillna(0, inplace=True)  # Replace NaN with zero

# Step 3: Prepare Data for LSTM
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# Create sequences
def create_sequences(data, seq_length):
    sequences = []
    for i in range(len(data) - seq_length):
        sequences.append(data[i:i+seq_length])
    return np.array(sequences)

seq_length = 60
X = create_sequences(scaled_data, seq_length)
y = scaled_data[seq_length:, 3]  # Predicting the close price

# Split data into training and test sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Step 4: Build LSTM Model
model = Sequential()
model.add(LSTM(units=64, return_sequences=True, input_shape=(seq_length, X.shape[2])))
model.add(Dropout(0.2))
model.add(LSTM(units=64))
model.add(Dropout(0.2))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')

# Step 5: Train the Model
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)

# Step 6: Make Predictions
predictions = model.predict(X_test)
predictions = scaler.inverse_transform(np.concatenate((np.zeros((predictions.shape[0], scaled_data.shape[1] - 1)), predictions), axis=1))[:, -1]

# Plot predictions
plt.figure(figsize=(14, 5))
plt.plot(data.index[train_size + seq_length:], predictions, color='red', label='Predicted Price')
plt.plot(data.index[train_size + seq_length:], data['Close'].values[train_size + seq_length:], color='blue', label='Actual Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# Step 7: Save and Deploy the Model
model.save('forex_lstm_model.h5')
```
